```python
def part_1(in_list):
    import re
    
    ## Inital read
    df = pd.Series(in_list).str.split(':',expand=True)
    df.columns = ['game', 'outcomes']
    
    ## Clean game convert to int
    df["game"] = df['game'].str.split(" ").apply(lambda x: int(x[1]))

    ## Clean ouctomes to dicts
    def string_to_dict_list (s):
        string_list = s.split(";")
        res = []
        for result in string_list:
            sub_dict = {}
            for pull in result.split(','):
                search_res =re.search(r"\s*(\d+)\s*([a-zA-Z]+)", pull)
                sub_dict[search_res[2]] = int(search_res[1])
            res.append(sub_dict)
        return res
    
    df["outcomes"] = df["outcomes"].apply(string_to_dict_list)

    def check_outcomes(dict_list):
        for d in dict_list:
            for k in bag.keys():
                if d.get(k,0) > bag[k]:
                    
                    ## BREAKING LOOP DON'T NEEED TO LOOK MORE
                    return False

        ## IF we didn't break the look we return True
        return True

    df['possible'] = df['outcomes'].apply(check_outcomes)

    def min_cubes(dict_list):
        min_dict = {}
        for d in dict_list:
            for k,v in d.items():
                if k in min_dict.keys():
                    min_dict[k] = max(v, min_dict[k])
                else:
                    min_dict[k] = v
        return min_dict

    df['min_cubes'] = df['outcomes'].apply(min_cubes)

    def power_calc(min_dict):
        res = 1
        for k,v in min_dict.items():
            res *= v
        return res

    df['power'] = df['min_cubes'].apply(power_calc)

    power_sum = df['power'].sum()
    
    return (df, power_sum)


## EXAMPLE
print("="*5 + " Example " + "="*20)
res_df, res_sum = part_1(example_input)
print(res_sum)
display(res_df)


## REAL
print("\n"*3)
print("="*5 + " Full " + "="*20)
res_df, res_sum = part_1(real_input)
print(res_sum)
display(res_df)
```
